## Question Three: Move All Zeros to End
**Description**  
Given a list of integers, move all zeros to the end while maintaining the relative order of the non-zero elements. 

**Examples**
```python
Input: nums = [0, 1, 0, 3, 12]
Output: [1, 3, 12, 0, 0]
```
```python
Input: nums = [4, 0, 5, 0, 0, 6]
Output: [4, 5, 6, 0, 0, 0]
```


````python
from typing import List
nums = [0, 1, 0, 3, 12]
def move_zeros_to_end(nums: List[int]) -> List[int]:
    write = 0
    for read in range(len(nums)):
        if nums [read] != 0:
            nums[write] = nums[read]
            write += 1

    for i in range(write, len (nums)):
        nums[i] = 0
    
    return nums

print(move_zeros_to_end(nums))

test = [0, 4, 6, 0, 9, 0, 0, 3]

move_zeros_to_end(test)

````


-   Explain why your solution works


The ADT I used is a list/array which achieves O(n) time complexity because it traverses the array only once. The first time to take note of all none-zeros and move non-zero elements forward, then filling remaining positions with zeros. Space complexity is O(1) since the rearrangement happens in-place without creating additional data structures. Alternative ADTs like queues would require O(n) additional space to separate elements, making them less efficient for this particular problem.

Answer: Time complexity: O(n), the problem asks us to shift positions of elements in a list; however, if we pop() an element, the entire list shifts which would force us to do a few iterative loops which would increase the time required by n.
